## ***1. Know Your Data***

### Import Libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
from google.colab import files
import pandas as pd

uploaded = files.upload()

titles = None
credits = None

for file in uploaded.keys():
    name = file.lower()
    if "title" in name:
        titles = pd.read_csv(file)
    elif "credit" in name:
        credits = pd.read_csv(file)

print("Titles shape:", titles.shape)
print("Credits shape:", credits.shape)
````

### Dataset First View

````python
data = pd.merge(titles, credits, on="id", how="left")
data.head()
````

### Variables Description

The dataset includes a mix of variables that describe different aspects of the content. The 'type' variable indicates whether the content is a Movie or a TV Show. The 'title' and 'description' provide basic information about the content.

Numerical variables such as 'release_year', 'runtime', 'imdb_score', and 'tmdb_score' help analyze trends, duration, and content quality. 'imdb_votes' and 'tmdb_popularity' give an idea of how popular a piece of content is among viewers.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
data = data.dropna(subset=["description"])

data["age_certification"] = data["age_certification"].fillna(data["age_certification"].mode()[0])

data["seasons"] = data["seasons"].fillna(0)

data = data.dropna(subset=["imdb_id"])

data["imdb_score"] = data["imdb_score"].fillna(data["imdb_score"].mean())
data["imdb_votes"] = data["imdb_votes"].fillna(0)
data["tmdb_score"] = data["tmdb_score"].fillna(data["tmdb_score"].mean())
data["character"] = data["character"].fillna("Unknown")

data = data.drop_duplicates().reset_index(drop=True)

data["content_age"] = 2024 - data["release_year"]

data.head()
````

### What all manipulations have you done and insights you found?

To prepare the dataset for analysis, several preprocessing steps were performed. First, the two datasets were merged using a common identifier to create a unified dataset. Then, missing values were handled using appropriate methods such as replacing with mean or mode, or removing entries where necessary.

Duplicate records were identified and removed to improve data quality. Some columns were cleaned and transformed to ensure consistency, and new features were created where needed to support better analysis.

After cleaning, the dataset became more reliable and easier to work with. Initial observations showed that movies are more common than TV shows, the data is heavily influenced by a few popular genres, and most of the content is relatively recent. These early insights helped guide further analysis.

#### Chart 1 - **Distribution of Content Type**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

sns.set_style("whitegrid")
sns.set_context("talk")

color_map = {
    "MOVIE": "#4C72B0",
    "SHOW": "#DD8452"
}

plt.figure(figsize=(7,4))
ax = sns.countplot(data=data, x="type", palette=color_map)

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x()+0.2, p.get_height()+50))

plt.title("Content Type Distribution", fontweight="bold")
plt.show()
````
