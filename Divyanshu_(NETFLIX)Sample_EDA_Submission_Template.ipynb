## 3. ***Data Wrangling***

### Data Wrangling Code

````python
data = data.drop_duplicates().reset_index(drop=True)

for col in data.columns:
    if data[col].dtype == "object":
        mode_value = data[col].mode()
        if len(mode_value) > 0:
            data[col] = data[col].fillna(mode_value[0])
    else:
        data[col] = data[col].fillna(data[col].median())

data["duration_num"] = data["duration"].str.extract("(\d+)").astype(float)
data["date_added"] = pd.to_datetime(data["date_added"], errors="coerce")

cat_cols = ["type", "rating", "country"]
num_cols = ["release_year", "duration_num"]

data.head()
````

### What all manipulations have you done and insights you found?

To prepare the dataset for analysis, duplicate rows were removed so that repeated observations would not distort the findings. Missing values were then handled according to the type of variable. For categorical features, missing entries were replaced using the most frequent value, while for numerical columns the median was used because it is less affected by extreme values.

The columns were also separated into categorical and numerical groups to make the analysis more organized. This made it easier to apply the right type of visualization and summary methods for each variable.

These preprocessing steps improved the quality and consistency of the dataset. The main insight from this stage is that data cleaning is essential before interpretation, because even a strong dataset can produce weak conclusions if quality issues are ignored.
